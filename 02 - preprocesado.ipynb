# Rellenar valores null en la columna

Primero verificamos los valores null

```python
print(df.isnull().sum()) # Verificacion numerica de valores null en toda la tabla

df[df["F_ESTRATOVIVIENDA"].isnull()] # verificacion en tabla de valores null FAMI_ESTRATOVIVIENDA
```

Se observa que tenemos 32137 valores null para la columna F_ESTRATOVIVIENDA


```python
df["F_ESTRATOVIVIENDA"].unique() #Valores en columna
```

```python
df['F_ESTRATOVIVIENDA'].head(50)
```

Rellenamos los valores null de acuerdo a la moda, ya que es un valor categorico

```python
df['F_ESTRATOVIVIENDA'] = df['F_ESTRATOVIVIENDA'].fillna(df['F_ESTRATOVIVIENDA'].mode()[0]) # Rellenamos los valores null con la moda

df['F_ESTRATOVIVIENDA'].head(50)
```

Finalmente podemos observar que en los valores 21 y 49 que anteriormente eran valores null, ahora tienen valores de acuerdo a la moda

```python
print(df.isnull().sum())
```

# Codificación Columna F_ESTRATOVIVIENDA (One-Hot Encoding)

```python
df["F_ESTRATOVIVIENDA"].unique() #Valores en columna
```

```python
df["F_ESTRATOVIVIENDA_encoded"] = pd.Categorical(
    df["F_ESTRATOVIVIENDA"],
    categories=["Sin Estrato", "Estrato 1", "Estrato 2", "Estrato 3", "Estrato 4", "Estrato 5", "Estrato 6"],
    ordered=True,
).codes

df[["F_ESTRATOVIVIENDA", "F_ESTRATOVIVIENDA_encoded"]].drop_duplicates().sort_values("F_ESTRATOVIVIENDA_encoded")
```
